```python
import pandas as pd
from sqlalchemy import create_engine
pd.options.display.float_format = '{:.2f}'.format
```

```python
csv_file = "Resources/WEOApr2018all.csv"
gdp_data_df = pd.read_csv(csv_file)
gdp_data_df.head()
```

```python
new_gdp_data_df = gdp_data_df[['WEO Country Code', 'WEO Subject Code', 'ISO', 'Country', 'Units', 'Scale', '2016']].copy()
new_gdp_data_df.head()
```

```python
new_ngdpd_data_df = new_gdp_data_df[new_gdp_data_df['WEO Subject Code'] == 'NGDPD']
new_ngdpd_data_df
```

```python
new_ngdpd_data_df = new_ngdpd_data_df[['ISO', 'Units', 'Scale', '2016']]
new_ngdpd_data_df
```

```python
new_ngdpd_data_df.columns = ['Country_code', 'Units', 'Scale', 'GDP_2016_USDollars']
new_ngdpd_data_df.dropna()

```

```python
rds_connection_string = "root:root@127.0.0.1/etl_db"
engine = create_engine(f'mysql+pymysql://{rds_connection_string}')
```

```python
new_ngdpd_data_df.to_sql(name='gdp_dollars_2016',con=engine,if_exists='append', index=False)
```

```python
bucket = 0
for i in range(len(new_ngdpd_data_df)):
    bucket +=1 
    new_ngdpd_data_df.iloc[i:i+1].to_sql(name='gdp_dollars_2016',con=engine,if_exists='append', index=False)
    print(bucket)
```
