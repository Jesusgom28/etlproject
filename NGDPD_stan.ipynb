```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/WEOApr2018all.csv"
gdp_data_df = pd.read_csv(csv_file)
gdp_data_df.head()
```

```python
new_gdp_data_df = gdp_data_df[['WEO Country Code', 'WEO Subject Code', 'Country', 'Units', 'Scale', '2016']].copy()
new_gdp_data_df.head()
```

```python
new_ngdpd_data_df = new_gdp_data_df[new_gdp_data_df['WEO Subject Code'] == 'NGDPD']
new_ngdpd_data_df
```

```python
new_ngdpd_data_df = new_ngdpd_data_df[['WEO Country Code', 'Country', 'Units', 'Scale', '2016']]
new_ngdpd_data_df
```

```python
new_ngdpd_data_df.columns = ['Country_id', 'Country', 'Units', 'Scale', 'NGDPD_2016']
new_ngdpd_data_df
```

```python
rds_connection_string = "root:root@127.0.0.1/ngdpd2016_db"
engine = create_engine(f'mysql+pymysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
new_ngdpd_data_df.to_sql(name='ngdpd2016_table', con=engine, if_exists='append', index=False)
```
